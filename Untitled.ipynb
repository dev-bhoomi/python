```python
#Program to Print Duplicates from a List of Integers in Python
a = [1, 2, 3, 1, 2, 4, 5, 6, 5]
seen=[]
duplicate=[]
for i in a:
    if i in seen:
        duplicate.append(i)
    else:
        seen.append(i)
print("duplicate",duplicate)
print("seen", seen)
```

```python
#find Cumulative sum of a list  means creating a new list where each element is the sum of all the previous elements
a = [1, 2, 3, 4]
total=0
result=[]
for i in a:
    total=total+i
    result.append(total)
print(result)
```

```python
#Break a List into Chunks of Size N in Python
a = [1, 2, 3, 4, 5, 6, 7, 8]
n=3
res=[]
for i in range(0, len(a),n):
    res.append(a[i:i+n])
print(res)
            
```
