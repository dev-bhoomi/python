```python
#Program to Check if a String is Palindrome or Not
s = "malayalam"
left = 0
right = len(s) - 1
is_palindrome = True

while left < right:
    if s[left] != s[right]:
        is_palindrome = False
        break
    left += 1          
    right -= 1       

if is_palindrome:
    print("Palindrome")
else:
    print("Not a palindrome")
```

```python
#Correct Way to Find Largest Element
arr = [1, 2, 3, 2, 3, 4, 20]

largest = arr[0]  # Assume first element is the largest

for num in arr:
    if num > largest:  # Only update if num is greater
        largest = num

print(largest)
```

```python
#Reverse Words in a Given String in Python
s = "Python is fun"
words=s.split()  ## words = ['Python', 'is', 'fun']
res=" "
for i in range(len(words)-1,-1,-1):
    res= res+ words[i]+ " "

res=res.strip()
print(res)
```
